# (a) Sketch the observations.

```python
X = np.transpose(np.array([[3,2,4,1,2,4,4],[4,2,4,4,1,3,1]]))
y = [0,0,0,0,1,1,1]

custom_colors = ['red','blue']
cmap = mcolors.ListedColormap(custom_colors)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap)
plt.xlabel('X_1')
plt.ylabel('X_2')
plt.show()
```

# (d) On your sketch, indicate the margin for the maximal margin hyperplane.

```python
x_plot = np.linspace(np.min(X[:, 0]), np.max(X[:, 0]), 100)
y_plot = x_plot- 0.5
plt.plot(x_plot, y_plot, 'k-')

y_plot = x_plot- 0
plt.plot(x_plot, y_plot, '--')

y_plot = x_plot- 1
plt.plot(x_plot, y_plot, '--')

custom_colors = ['red','blue']
cmap = mcolors.ListedColormap(custom_colors)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap)
plt.xlabel('X_1')
plt.ylabel('X_2')
plt.show()
```

# (e) Indicate the support vectors for the maximal margin classifier.

```python
# support vectors:
SVs = (1,2,4,5)
X_SVs = np.array([X[i] for i in SVs])
y_SVs = [y[i] for i in SVs]
```

```python
x_plot = np.linspace(np.min(X[:, 0]), np.max(X[:, 0]), 100)
y_plot = x_plot- 0.5
plt.plot(x_plot, y_plot, 'k-')

x_plot = np.linspace(np.min(X[:, 0]), np.max(X[:, 0]), 100)
y_plot = x_plot- 0
plt.plot(x_plot, y_plot, '--')

x_plot = np.linspace(np.min(X[:, 0]), np.max(X[:, 0]), 100)
y_plot = x_plot- 1
plt.plot(x_plot, y_plot, '--')

custom_colors = ['red','blue']
cmap = mcolors.ListedColormap(custom_colors)
plt.scatter(X_SVs[:, 0], X_SVs[:, 1], c=y_SVs, cmap=cmap)
plt.xlabel('X_1')
plt.ylabel('X_2')
plt.show()
```
